```python
# Load the model.
model = YOLO('yolov8n.pt')

# Training.
results = model.train(
   data='/content/WR-Finder-3/data.yaml',
   imgsz=640,
   epochs=10,
   batch=8,
   name='yolov8n_WR')
```
